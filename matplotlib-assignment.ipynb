```python
# 4. Using the given x and y data, generate a 3D surface plot to visualize the function 

# x = np.linspace(-5, 5, 100)
# y = np.linspace(-5, 5, 100)
# x, y = np.meshgrid(x, y)
# z = np.sin(np.sqrt(x**2 + y**2))
# data = {
#     'X': x.flatten(),
#     'Y': y.flatten(),
#     'Z': z.flatten()
# }
# df = pd.DataFrame(data)

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Provided dataset
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create a 3D surface plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surf = ax.plot_surface(x, y, z, cmap='viridis')

# Add labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title('3D Surface Plot of z = sin(sqrt(x^2 + y^2))')

# Add color bar
fig.colorbar(surf)

plt.show()
```

```python


# 5. Create a Bokeh heatmap using the provided dataset.

# np.random.seed(25)
# data = {
#     'Country': ['USA', 'Canada', 'UK',
# 'Germany', 'France'],
#     'Population':
# np.random.randint(100, 1000, 5),
#     'GDP': np.random.randint(500, 2000,
# 5)
# }
# df = pd.DataFrame(data)
import numpy as np
import pandas as pd
from bokeh.io import output_file, show
from bokeh.models import ColorBar
from bokeh.plotting import figure
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256

# Provided dataset
np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

# Bokeh Heatmap
output_file("heatmap.html")

# Define colormap
colors_population = linear_cmap(field_name='Population', palette=Viridis256 ,low=min(df['Population']), high=max(df['Population']))
colors_gdp = linear_cmap(field_name='GDP', palette=Viridis256 ,low=min(df['GDP']), high=max(df['GDP']))

# Create a figure
p = figure(plot_width=800, plot_height=300, title="Country Data Heatmap", x_range=df['Country'], y_range=['Population', 'GDP'])

# Add the heatmap rectangles for Population
p.rect(x='Country', y='Population', width=0.9, height=0.2, source=df, fill_color=colors_population, line_color=None, legend_label='Population')

# Add the heatmap rectangles for GDP
p.rect(x='Country', y='GDP', width=0.9, height=0.2, source=df, fill_color=colors_gdp, line_color=None, legend_label='GDP')

# Add color bars
color_bar_population = ColorBar(color_mapper=colors_population['transform'], width=8, location=(0,0))
color_bar_gdp = ColorBar(color_mapper=colors_gdp['transform'], width=8, location=(0,0))

p.add_layout(color_bar_population, 'right')
p.add_layout(color_bar_gdp, 'right')

# Adjust the legend location
p.legend.location = "top_left"

# Show the plot
show(p)
```

```python
#1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
import numpy as np
from bokeh.plotting import figure, output_file, show

# Generate data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Output to HTML file (optional)
output_file("sine_wave.html")

# Create a new plot with a title and axis labels
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=800, height=400)

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label="sin(x)", line_width=2)

# Show the result
show(p)
```

```python
# 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
# markers based on the 'sizes' and 'colors' columns.
import numpy as np
import pandas as pd
from bokeh.plotting import figure, output_file, show

# Generate random data
np.random.seed(0)
data = {
    'x': np.random.rand(100) * 10,  # Random x values from 0 to 10
    'y': np.random.rand(100) * 10,  # Random y values from 0 to 10
    'sizes': np.random.rand(100) * 30 + 5,  # Random marker sizes from 5 to 35
    'colors': np.random.rand(100),  # Random marker colors
}

# Convert data to DataFrame
df = pd.DataFrame(data)

# Output to HTML file (optional)
output_file("scatter_plot.html")

# Create a new plot with a title
p = figure(title="Bokeh Scatter Plot", width=800, height=400)

# Add scatter plot with different marker sizes and colors
p.scatter('x', 'y', size='sizes', color='colors', alpha=0.6, source=df)

# Add axis labels
p.xaxis.axis_label = "X"
p.yaxis.axis_label = "Y"

# Show the result
show(p)
```

```python
# 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
# fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
# counts = [20, 25, 30, 35]
from bokeh.plotting import figure, output_file, show
from bokeh.models import ColumnDataSource

# Provided dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts, color=['red', 'orange', 'yellow', 'green']))

# Output to HTML file (optional)
output_file("bar_chart.html")

# Create a new plot with a title and axis labels
p = figure(x_range=fruits, title="Fruit Counts", plot_height=350, plot_width=600, toolbar_location=None, tools="")

# Add vertical bars
p.vbar(x='fruits', top='counts', width=0.9, color='color', legend_field="fruits", source=source)

# Add axis labels
p.xaxis.axis_label = "Fruits"
p.yaxis.axis_label = "Counts"

# Set the y range explicitly to avoid clipping of bars
p.y_range.start = 0

# Remove gridlines
p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None

# Show the result
show(p)
```

```python
# 4. Create a Bokeh histogram to visualize the distribution of the given data.
# data_hist = np.random.randn(1000)
# hist, edges = np.histogram(data_hist, bins=30)
import numpy as np
from bokeh.plotting import figure, output_file, show

# Given data
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Output to HTML file (optional)
output_file("histogram.html")

# Create a new plot with a title and axis labels
p = figure(title="Histogram", width=800, height=400)
p.xaxis.axis_label = 'Value'
p.yaxis.axis_label = 'Frequency'

# Add the histogram bars
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white", alpha=0.5)

# Show the result
show(p)

```

```python
# 5.Create a Bokeh heatmap using the provided dataset.
# data_heatmap = np.random.rand(10, 10)
# x = np.linspace(0, 1, 10)
# y = np.linspace(0, 1, 10)
# xx, yy = np.meshgrid(x, y)
import numpy as np
from bokeh.plotting import figure, output_file, show
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar
from bokeh.palettes import Viridis256

# Provided dataset
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Output to HTML file (optional)
output_file("heatmap.html")

# Flatten the data for ColumnDataSource
data = {'x': xx.ravel(),
        'y': yy.ravel(),
        'value': data_heatmap.ravel()}

# Create a ColumnDataSource
source = ColumnDataSource(data)

# Create a color mapper
color_mapper = linear_cmap(field_name='value', palette=Viridis256 ,low=min(data['value']), high=max(data['value']))

# Create a new plot with a title
p = figure(title="Heatmap", x_range=(0, 1), y_range=(0, 1), plot_width=800, plot_height=400)

# Add the heatmap
p.rect(x='x', y='y', width=1/10, height=1/10, source=source, fill_color=color_mapper, line_color=None)

# Add color bar
color_bar = ColorBar(color_mapper=color_mapper['transform'], width=8, location=(0,0))
p.add_layout(color_bar, 'right')

# Show the result
show(p)
```
